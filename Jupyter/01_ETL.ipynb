## Nulos

Se verifica si hay datos nulos

```python
functions.tipo_dato(victimas)
```

Se observa que la columnas de Edad y fecha de fallecimiento tienen varios tipos de datos. Por lo tanto se revisan para asigrnar un solo tipo de dato

```python
#Se observa primero "Edad".
victimas["Edad"].apply(type).value_counts()
```

Son 53 datos en str, esto sucede ya que no se tiene el dato de la "Edad", por que se decide remplazarlos por la edad promedio de fallecimiento. Para hacer esto se compara tambien el "Sexo", "Rol" y "Victima". remplazando en caso de que falte por el mas comun.

```python
victimas[victimas["Edad"].apply(lambda x: isinstance (x,str))]
```

```python
# Se reemplaza "SD" con NaN en la columna
victimas["Sexo"].replace("SD", pd.NA,inplace= True)
# Se calcula el valor mas comun en la columna
moda = victimas["Sexo"].mode().iloc[0]
print(f"El valor mas comun es {moda}")
#Se remplazan los valores Nulos con el valor mas frecuente.
victimas["Sexo"].fillna(moda,inplace=True)
```

Una vez que remplazamos el "Sexo" con "MASCULINO" por ser el valor mas comun, se remplaza la edad mas comun segun el sexo.

```python
#Se remplaza "SD" con Nan en la columna
victimas["Edad"].replace("SD", pd.NA, inplace= True)
#Se calcula el promedio de edad segun el "Sexo"
edad_sexo = victimas.groupby("Sexo")["Edad"].mean()
print(f'La edad promedio de Masculino es {round(edad_sexo["MASCULINO"])}')
print(f'La edad promedio de Femenino es {round(edad_sexo["FEMENINO"])}')
#Se remplazan los nulos por el promedio de edad segun el sexo.
victimas["Edad"] = victimas.apply(lambda row: edad_sexo[row["Sexo"]] if pd.isna(row["Edad"]) else row["Edad"],axis =1)
#Se convierte a entero.
victimas["Edad"] = victimas["Edad"].astype(int)
```

Ahora se revisan las columnas de "Rol" y "Victima"

```python
#Cantidad de datos faltantes en columna "Rol"
len(victimas[victimas["Rol"]== "SD"])
```

```python
#Cantidad de datos faltantes en columna "Victima"
len(victimas[victimas["Victima"]== "SD"])
```

Son 11 y 9 datos que faltan en las columnas de "Rol" y "Victima" respectivamente, se las remplaza por los valores mas comunes.

```python
# Se reemplaza "SD" con NaN en la columna
victimas["Rol"].replace("SD", pd.NA,inplace= True)
# Se calcula el valor mas comun en la columna
moda = victimas["Rol"].mode().iloc[0]
print(f"El valor mas comun es {moda}")
#Se remplazan los valores Nulos con el valor mas frecuente.
victimas["Rol"].fillna(moda,inplace=True)
```

```python
# Se reemplaza "SD" con NaN en la columna
victimas["Victima"].replace("SD", pd.NA,inplace= True)
# Se calcula el valor mas comun en la columna
moda = victimas["Victima"].mode().iloc[0]
print(f"El valor mas comun es {moda}")
#Se remplazan los valores Nulos con el valor mas frecuente.
victimas["Victima"].fillna(moda,inplace=True)
```

La columna "Fecha Fallecimiento", no aporta informacion util para el objetivo de reducir la cantidad de fallecimientos en siniestros viales, por lo tanto se decide eliminarla.
